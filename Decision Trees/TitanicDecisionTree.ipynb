```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

# Converting male and female into numerical values.
d = {'male': 0, 'female': 1}
df['Sex'] = df['Sex'].map(d)

# Any data that is missing from Age in the dataset will be filled in with a median age of passangers, to prevent missing data
df['Age'].fillna(df['Age'].median(), inplace=True)

# The Pclass, Sex and Age columns will be used for Predicting survival
features = ['Pclass', 'Sex', 'Age']
X = df[features]
y = df['Survived']

dtree = DecisionTreeClassifier(max_depth=3)
dtree.fit(X, y)
plt.figure(figsize=(15, 15))
tree.plot_tree(dtree, feature_names=features, class_names=['Not Survived', 'Survived'], filled=True, fontsize= 10) # Added font size, this allows the tree to be more visable
plt.title("Decision Tree for Titanic Survival Prediction")
plt.tight_layout()
plt.show()
```
