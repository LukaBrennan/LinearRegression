# Understanding the decision tree. 
- The Gender Field: The decision tree has to chose a midpoint. In my code I have the gender set to a numerical value. 0 representing male and female representing 1. the decision tree finds the midpoint of this which is 0.5 so if you are male i.e <0.5 toy will move to the True. 
# What are the main parts.
- I have the male and female parameter set to 0 and 1. the reason for this is that in the decision tree, it will check if 0 is <=0.5, what this is doing is checking in the decision tree if the person was either male or female. 
- this is populated on the decision tree as the name "Sex" which his mapped to the decision tree. 
- I have a fillna method. what this does is it fill in any missing Age filed in the .csv file. the reason for this is to prevent any inconstant data, because in my understanding having empty data filed in with average data will be better than not having it. 
- The "features" which has Pclass, Sex and age, are all populated to the decision tree. 
- Why is there a max depth?. the reason for this is that is makes the decision tree readable. how?. The deeper the tree the more complex the model, so what this means having it set to 3 will reduce overfitting and by that reduced the size of the decision tree. 
- Figure size. this just changes the size of the decision tree. Having it set to (15, 15) i believe is the sweet spot. 
- Plot tree. This is a visualization tool used to display the decision tree with all of it's features. such as dtree which holds the x and y values. 

```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")

# Converting male and female into numerical values.
d = {'male': 0, 'female': 1}
df['Sex'] = df['Sex'].map(d)

# Any data that is missing from Age in the dataset will be filled in with a median age of passangers, to prevent missing data
df['Age'].fillna(df['Age'].median(), inplace=True)

# The Pclass, Sex and Age columns will be used for Predicting survival
features = ['Pclass', 'Sex', 'Age']
X = df[features]
y = df['Survived']

dtree = DecisionTreeClassifier(max_depth=3)
dtree.fit(X, y)
plt.figure(figsize=(15, 15))
tree.plot_tree(dtree, feature_names=features, class_names=['Survived', 'Not Survived'], filled=True, fontsize= 10) # Added font size, this allows the tree to be more visable
plt.title("Decision Tree for Titanic Survival Prediction")
plt.tight_layout()
plt.show()
```
